```python
import pandas as pd
import matplotlib
%matplotlib inline
matplotlib.rcParams['figure.figsize']=[12,8]
```

```python
data_set=pd.read_csv('C:/Users/AMRIT/Desktop/PROJECT CERTI/TIME FORECASTING/students.csv')
data_set.head(10)
data_set.isnull().sum()
data_set.dtypes
```

```python
data_set['Month']=pd.to_datetime(data_set['Month'])
data_set.head()
data_set.dtypes
data_set=data_set.set_index('Month')
data_set.head()
data_set.plot(grid=True)
matplotlib.rcParams['figure.figsize']=[12.0,8.0]
from datetime import datetime
start_date=datetime(2019,1,1)
end_date=datetime(2019,12,30)
data_set[(start_date <=data_set.index) & (data_set.index <= end_date)].plot(grid=True)
```
