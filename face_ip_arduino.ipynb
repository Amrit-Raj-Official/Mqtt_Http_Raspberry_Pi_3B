```python
import numpy as np
import serial
import time
import sys
import cv2
arduino = serial.Serial(port='COM7',baudrate=9600) 
time.sleep(2)
print("Connected to arduino...")
import requests
```

```python
cascPath ='C:/Users/AMRIT/Desktop/haarcascade_frontalface_default.xml'
print("enter ip with port no.:")
i=input()
faceCascade = cv2.CascadeClassifier(cascPath)
while True:
    url="http://{}/shot.jpg".format(i)
    data=requests.get(url)
    myphoto=data.content    
    myphoto_a=bytearray(myphoto)
    photor=np.array(myphoto_a)
    photo_3d=cv2.imdecode(photor,-1)
    gray = cv2.cvtColor(photo_3d, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5,minSize=(30, 30),)
    for (x, y, w, h) in faces:
        cv2.rectangle(photo_3d, (x, y), (x+w, y+h), (0, 255, 0), 2)
        roi_gray  = gray[y:y+h, x:x+w]
        roi_color = photo_3d[y:y+h, x:x+w]
        arr = {y:y+h, x:x+w}
        cv2.circle(photo_3d, (int((2*x+w)/2),int((2*y+h)/2)), 5, (255, 255, 255), -1)

    cv2.imshow("Faces found", photo_3d)
    if (len(faces)) != 0:
        arduino.write('Y'.encode())
        print ("Face is detected")
    if (len(faces)) == 0:
        arduino.write('n'.encode())
        print ("Face is not detected")
    if cv2.waitKey(1)==13:
        break
cv2.destroyAllWindows()
```
